```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
archivo = 'C:/Programaci칩n/Lab Intermedio/Doble rendija/Datos-Bombillo-una rendija.xlsx'

# Cargar los datos desde el archivo Excel
df = pd.read_excel(archivo)

# Agrupar por la columna 'Distancia', calcular el promedio y la desviaci칩n est치ndar de 'Fotones'
agrupado = df.groupby('Distancia')['Fotones'].agg(['mean', 'std']).reset_index()
agrupado.columns = ['Distancia', 'Promedio_Fotones', 'Desviacion_Estandar']

# Crear el gr치fico
plt.figure(figsize=(8, 6))
plt.scatter(
    agrupado['Distancia']-3000, 
    agrupado['Promedio_Fotones'], label='Fotones Promedio'
    
)
plt.xlabel('Distancia($\mu m$)')
plt.ylabel('Promedio de Fotones por Segundo')
plt.title('Distancia vs Fotones Promedio por Segundo para una rendija')
plt.grid(True)
plt.legend()
plt.show()

# Guardar el resultado en un array
resultados = agrupado.to_numpy()
#print(resultados)
```

```python
archivo = 'C:/Programaci칩n/Lab Intermedio/Doble rendija/Datos-Bombillo-una rendija.xlsx'


df = pd.read_excel(archivo)


agrupado = df.groupby('Distancia')['Fotones'].agg(['mean', 'std']).reset_index()
agrupado.columns = ['Distancia', 'Promedio_Fotones', 'Desviacion_Estandar']


plt.figure(figsize=(8, 6))
plt.errorbar(
    agrupado['Distancia']-3000, 
    agrupado['Promedio_Fotones'], 
    yerr=agrupado['Desviacion_Estandar'], 
    fmt='o', 
    capsize=5, 
    label='Fotones Promedio',
    ecolor='red',  
    elinewidth=2   
)
plt.xlabel('Distancia($\mu m$)')
plt.ylabel('Promedio de Fotones por Segundo')
plt.title('Distancia vs Fotones Promedio por Segundo para una rendija')
plt.grid(True)
plt.legend()
plt.show()


resultados = agrupado.to_numpy()
#print(resultados)
```

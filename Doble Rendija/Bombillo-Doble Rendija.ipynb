# Gr√°ficas Rendija Doble-Bombillo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


archivo = 'C:/Programaci√≥n/Lab Intermedio/Doble rendija/Datos_bombillo.xlsx'  
df = pd.read_excel(archivo)


agrupado = df.groupby('x (+- 10 mu m )')['promedio'].agg(['mean', 'std']).reset_index()
agrupado.columns = ['x (+- 10 mu m )', 'promedio', 'Desviacion_Estandar']


plt.figure(figsize=(8, 6))
plt.scatter(agrupado['x (+- 10 mu m )']-1550, agrupado['promedio'], label='Fotones Promedio',s=5)

plt.xlabel('Distancia($\mu m$)')
plt.ylabel('Promedio de Fotones por Segundo')
plt.title('Distancia vs Fotones Promedio por Segundo-Doble Remdija')
plt.grid(True)
plt.legend()
plt.show()


resultados = agrupado.to_numpy()
#print(resultados)
```

```python
datos = pd.read_excel(archivo, sheet_name='Sheet1')

valores_x = datos['x (+- 10 mu m )']
valores_promedio = datos['promedio']


desviacion_estandar = datos[['conteo 1 (+-1)', 'conteo 2 (+-1)', 'conteo 3 (+-1)']].std(axis=1)


plt.figure(figsize=(10, 6))
plt.errorbar(valores_x, valores_promedio, yerr=desviacion_estandar, fmt='o', ecolor='red', capsize=5, capthick=1, color='blue', label='Fotones Promedio')
plt.title('Gr√°fica de Posici√≥n vs Promedio de Fotones con Incertdumbres - Doble Rendija')
plt.xlabel('Distancia ($\mu m$)')
plt.ylabel('Promedio de Fotones por Segundo')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo 1
archivo_1 = 'C:/Programación/Lab Intermedio/Doble rendija/Datos-Láser-una rendija_01.xlsx'

# Cargar datos
datos_1 = pd.read_excel(archivo_1)

# Crear gráfico
plt.figure(figsize=(8, 6))
plt.plot(-datos_1['d (micro)']+6600, datos_1['V'], marker='o')
plt.xlabel('Distancia ($\mu m$)')
plt.ylabel('V')
plt.title('Gráfica de distancia ($\mu m$) vs Voltaje (V) para la primera rendija')
plt.grid(True)
#plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy.optimize import curve_fit

data = "C:/Programación/Lab Intermedio/Oscilador Torsional/Momento Inercial - Oscilador Torsional.xlsx"
data = pd.read_excel(data)
print(data)
```

```python
x,y= np.array(data['Cuadrantes de latón usados']), np.array(data['T (s)'])
```

```python
plt.scatter(x, y)
```

```python
param, cov= scipy.optimize.curve_fit(lambda t,a,b: a*t+b,  x,y)
incert =np.sqrt(np.diag(cov))
```

```python
param, incert
```

```python
X=np.linspace(x[0], x[-1], 1000)
Y=param[0]*X+param[1]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(X, Y, color='red', linewidth=2, 
        label=f'Linear fit: v = ({np.round(param[0], 3)}±{np.round(incert[0],3)})t + ({np.round(param[1],3)}±{np.round(incert[1],3)})')
#plt.legend(['Regresión Lineal: Intercepto '])
ax.legend(loc='upper left')
ax.set_xlabel('Cuadrantes')
ax.set_ylabel('T (s)')
ax.set_title('Regresión Lineal-Cuadrantes vs. T (s)')
```
